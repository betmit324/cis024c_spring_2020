```python
def convert_to_dt(val):
    
    dt_list = []
    
    for item in val:
        dt_list.append(item.date())
        
    return dt_list
```

## Bar Charts

```python
df_covid
```

```python
df_covid_dict = df_covid.sum().to_dict()
df_covid_dict

del df_covid_dict['date']
```

```python
del df_covid_dict['World']
```

```python
country = []
total_death = []

for key in df_covid_dict:
    
    if df_covid_dict[key] < 100:
        continue
    
    country.append(key)
    total_death.append(df_covid_dict[key])

country,total_death
```

```python
df_covid_dc = df_covid.transpose().reset_index().drop(df_covid.index[0:2])
df_covid_dc
```

```python
plt.clf()


plt.figure(figsize=(20,40))

plt.xlabel('country')
plt.ylabel('total deaths')
plt.title("Covid-19 deaths by country")

plt.barh(country, total_death, color='r')
plt.show()
plt.barh(country, np.log(total_death), color='b')
plt.show()
```

## Pie Charts


```python

sales = [10000,20000,30000,40000]
regions = ['North','South','East','West']

colors = ['r','g','b','y']

plt.title("My Sales per Region")
plt.legend()
plt.pie(sales, 
        labels=regions,  
        colors= colors, 
        shadow = True, 
        explode=(0,0.2,0,0), 
        autopct = '%1.1f%%',
        radius=1)
    
```

## Subplot

```python
x = np.linspace(1,100,num=100)
y = np.random.normal(10,2,100)
```

```python
plt.close('all')
```

```python
f, ax = plt.subplots()

ax.plot(x,y)
ax.set_title('simple plot')

plt.savefig('test.png')
```

```python
f, axarr = plt.subplots(2)

axarr[0].plot(x,y)
axarr[0].set_title('Two subplots')

axarr[1].hist(y)

plt.show()
```

```python

f, axarr = plt.subplots(1,2)

f.set_figheight(10)
f.set_figwidth(20)

axarr[0].plot(x,y)
axarr[0].set_title('Two subplots')
axarr[0].set_xlabel('some number')

axarr[1].hist(y)

plt.show()
```

```python

f, axarr = plt.subplots(2,2)

f.set_figheight(10)
f.set_figwidth(20)

axarr[0,0].plot(x,y)
axarr[0,0].set_title('Two subplots')
axarr[0,0].set_xlabel('some number')

axarr[0,1].hist(y)

axarr[1,0].plot(x,y)
axarr[1,0].set_title('Two subplots')
axarr[1,0].set_xlabel('some number')

axarr[1,1].hist(y)


plt.show()
```

## Seaborn test

```python
sns.set(style="ticks")

dots = sns.load_dataset("dots")

# Define a palette to ensure that colors will be
# shared across the facets
palette = dict(zip(dots.coherence.unique(),
                   sns.color_palette("rocket_r", 6)))

# Plot the lines on two facets
sns.relplot(x="time", y="firing_rate",
            hue="coherence", size="choice", col="align",
            size_order=["T1", "T2"], palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```
